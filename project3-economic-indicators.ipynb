# Nigeria Economic Indicators

**Data Source**: Data sourced from Central Bank of Nigeria (CBN), National Bureau of Statistics (NBS), and World Bank published records.

**Objective**: To analyse Nigeria's key economic indicators from 2015 to 2024 using Python, including inflation, exchange rate, GDP growth, oil prices, and foreign reserves.

**Tools**: Python, Pandas, Matplotlib

```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("ready")
```

```python
# Dataset

data = {
    "year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "inflation_pct": [9.0, 15.7, 16.5, 12.1, 11.4, 13.2, 17.0, 18.8, 24.5, 32.7],
    "exchange_rate_usd": [197, 253, 306, 306, 307, 381, 410, 422, 770, 1400],
    "gdp_growth_pct": [2.7, -1.6, 0.8, 1.9, 2.2, -1.8, 3.4, 3.3, 2.9, 3.2],
    "oil_price_usd": [52, 44, 54, 71, 64, 42, 70, 101, 83, 80],
    "foreign_reserves_bn": [29.1, 25.8, 38.8, 42.6, 38.0, 35.4, 40.5, 37.1, 32.9, 34.2]
}

# Inserting the dataset into a dat frame

df = pd.DataFrame(data)
print(df)
```

```python
print(df.shape)
print(df.dtypes)
print(df.describe())
```

```python
#Plotting

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Inflation
axes[0, 0].plot(df['year'], df['inflation_pct'], marker='o', color='red')
axes[0, 0].set_title('Inflation Rate (%)')
axes[0, 0].set_xlabel('Year')

# Exchange Rate
axes[0, 1].plot(df['year'], df['exchange_rate_usd'], marker='o', color='orange')
axes[0, 1].set_title('Exchange Rate (NGN/USD)')
axes[0, 1].set_xlabel('Year')

# GDP Growth
axes[1, 0].plot(df['year'], df['gdp_growth_pct'], marker='o', color='steelblue')
axes[1, 0].set_title('GDP Growth Rate (%)')
axes[1, 0].set_xlabel('Year')

# Oil Price
axes[1, 1].plot(df['year'], df['oil_price_usd'], marker='o', color='green')
axes[1, 1].set_title('Oil Price (USD)')
axes[1, 1].set_xlabel('Year')

plt.suptitle('Nigeria Economic Indicators 2015-2024')
plt.tight_layout()
plt.show()
```

```python
# Correlation between oil prices and GDP growth

correlation = df['oil_price_usd'].corr(df['gdp_growth_pct'])
print(correlation)
```

```python
print(f"Correlation between Oil price($) and the GDP growth (%): {correlation:.2f}")
```

```python
# Correlatio between inflation and exchange rate

correlation2 = df['inflation_pct'].corr(df['exchange_rate_usd'])
print(f"Correlation between inflation and exchange rate: {correlation2:.2f}")
```

# Findings

When the Naira weakens, the price of importation becomes more expensive, resulting in higher inflation values. A correlation of 0.94 aptly reflected this

The GDP growth of Nigeria is heavily tied to the price of oil, this was expressed in the positive correlation of 0.78 between oil prices and national GDP growth
